```python
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd
import regex as re
```

```python
akutagawa_works='https://www.aozora.gr.jp/index_pages/person879.html#sakuhin_list_1'
```

```python
page_a = rq.get(akutagawa_works)
print(page_a)
```

```python
page_a.encoding = 'utf-8'
print(page_a.text)
```

```python
soup = BeautifulSoup(page_a.text, features="html.parser")
```

```python
print(soup.prettify())
```

```python
# Необходимо собрать ссылку на каждое произведение, собираем ссылку по частям (сначала окончание ссылки, затем главная часть)
part_link = []
for i in soup.find_all('a'):
  if i.parent.name == 'li' and 'card' in i.get('href'):
    part_link.append(i.get('href'))

part_link
```

```python
corrected_part_link = [i.replace('..','') for i in part_link]
```

```python
# в список full_links собрали ссылки на все карточки произведений
main_part = 'https://www.aozora.gr.jp'
full_links = []

for link in corrected_part_link:
  link = main_part + link
  full_links.append(link)

full_links
```

Задача - собрать ссылки на все произведения

```python
all_works = []
for card in full_links:
  #url2 = full_links[0]
  page2 = rq.get(card)
  page2.encoding = 'utf-8'
  soup2 = BeautifulSoup(page2.text, features="html.parser")

  final_link = []
  for i in soup2.find_all('a'):
    if i.parent.name == 'td' and 'files' in i.get('href') and 'html' in i.get('href'):
      final_link.append(i.get('href'))

  for link in final_link:
      end_link = re.sub(r'^\.', '', link)
      link0 =re.sub(r'\/card\d*\.html', '', full_links[0])+end_link

  all_works.append(link0)
```

Задача - написать функцию, которая соберет информацию по всем произведениям.



```python
def parsed_data (url0):
  page0 = rq.get(url0)
  page0.encoding = 'Shift_JIS'
  soup0 = BeautifulSoup(page0.text, features="html.parser")
  title = soup0.find_all('h1')[0].text
  #author = soup0.find_all('h2')[0].text
  text = soup0.find_all('body')[0].text
  #final_text = ' '.join(text)
  #final_text = final_text.replace('\xa0', ' ')
  return url0, author, title, text
```

```python
corpus = []

for work in all_works:
  processed = parsed_data(work)
  corpus.append(processed)
```

```python
df = pd.DataFrame(corpus)
df.head(10)
```
